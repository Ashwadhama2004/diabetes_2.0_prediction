```python
X=dataset.drop("glyhb",axis=1)
X=dataset.drop("id",axis=1)
```

```python
X = pd.DataFrame(scaled_dataset, columns=X.columns)
```

```python
Y=dataset["glyhb"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
classo1=RandomForestRegressor()
```

```python
classo1.fit(X_train,Y_train)

```

```python
train_predictions=classo1.predict(X_train)
r2_score(train_predictions,Y_train)
```

```python
test_predictions=classo1.predict(X_test)
r2_score(test_predictions,Y_test)
```
