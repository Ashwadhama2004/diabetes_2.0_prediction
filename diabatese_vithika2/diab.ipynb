```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

```python
dataset = pd.read_csv("diabetes.csv")
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
dataset.drop(columns=["id","bp.2s","bp.2d"],axis=1, inplace=True)
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.dropna(inplace=True)

```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
dataset["location"].value_counts() 
```

```python
dataset['gender'].value_counts()
```

```python
dataset["frame"].value_counts()
```

```python
dataset["location"] = dataset["location"].replace({"Buckingham": 0, "Louisa": 1})
dataset["gender"] = dataset["gender"].replace({"female": 0, "male": 1})
dataset["frame"] = dataset["frame"].replace({"small": 0, "medium": 1, "large": 2})
```

```python
dataset.head()
```

```python
X = dataset.drop(columns=["glyhb"])
```

```python
X.head()
```

```python
X.describe()
```

```python
Y=dataset["glyhb"]
```

```python
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, Y_train)
```

```python

y_pred = model.predict(X_test)

# Evaluate
train_score = model.score(X_train, Y_train)
test_score = model.score(X_test, Y_test)

train_r2 = r2_score(Y_train, model.predict(X_train))
test_r2 = r2_score(Y_test, y_pred)

print(f"Model score (train): {train_score}, R^2 (train): {train_r2}")
print(f"Model score (test): {test_score}, R^2 (test): {test_r2}")
```

```python
input_data=(203,90,36.0,5.6,0,22,2,60,160.9,6,105,72,36,38,630)

input_df = pd.DataFrame([input_data])

# Make prediction
glyhb_prediction = model.predict(input_df)[0]
    
# Determine diabetes status
diabetes_status = "No diabetes" if glyhb_prediction < 6.5 else "Diabetes"

print(f"Predicted glyhb: {glyhb_prediction:.2f} - Status: {diabetes_status}")
```
